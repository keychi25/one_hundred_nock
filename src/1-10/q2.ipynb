```python
# Gray scale
def BGR2GRAY(img):
	# OpenCVでの変換
	# out = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
	# OpenCVでの変換
	b = img[:, :, 0].copy()
	g = img[:, :, 1].copy()
	r = img[:, :, 2].copy()

	# Gray scale
	out = 0.2126 * r + 0.7152 * g + 0.0722 * b
	out = out.astype(np.uint8)

	return out
```

```python
out = BGR2GRAY(img)
```

```python
plt.imshow(out, cmap = "gray")
```
