# BGRからRGBにへんかんする

```python
from pprint import pprint
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("data/imori.jpg")
```

```python
# 画像の表示(cv2で読み込むとBGRになる)
plt.imshow(img)
```

```python
def BGR2RGB(image): 
    blue = img[:, :, 0].copy()
    green = img[:, :, 1].copy()
    red = img[:, :, 2].copy()

    image[:, :, 0] = red
    image[:, :, 0] = green
    image[:, :, 0] = blue
    return image
```

```python
# 変換BGR2RGB(img)
```

```python
# 変換後の画像の表示
plt.imshow(image)
```
