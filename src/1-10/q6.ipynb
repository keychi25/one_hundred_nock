```python
# 値によって減色処理を施す
def subtractive_colo(pix):
    if  0 <= pix <  64:
        pix = 32
    elif 64 <= pix < 128:
        pix = 92
    elif 128 <= pix <192:
        pix = 160
    elif 192 <= pix < 256:
        pix = 224 
    elif pix < 0 or pix > 257:
        pix = 0
    return pix 
```

```python
# Dicrease color
def dicrease_color(img):
	out = img.copy()
	out = out // 64 * 64 + 32
	return out
```

```python
# Dicrease color
out = dicrease_color(img)
plt.imshow(out)
```
